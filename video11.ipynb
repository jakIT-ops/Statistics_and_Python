###       (     )


```python
def pmean_test(sample, mean0, p_var, alpha=0.05):
    s_mean = np.mean(sample)
    n = len(sample)
    rv = stats.norm()
    interval = rv.interval(1-alpha)

    z = (s_mean - mean0) / np.sqrt(p_var/n)
    if interval[0] <= z <= interval[1]:
        print('  ')
    else:
        print('   ')

    if z < 0:
        p = rv.cdf(z) * 2
    else:
        p = (1 - rv.cdf(z)) * 2
    print(f'p value {p:.3f}')
```

```python
pmean_test(sample, 130, 9)
```

###       


```python
def pvar_test(sample, var0, alpha=0.05):
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.chi2(df=n-1)
    interval = rv.interval(1-alpha)
    
    y = (n-1) * u_var / var0
    if interval[0] <= y <= interval[1]:
        print('   ')
    else:
        print('   ')

    if y < rv.isf(0.5):
        p = rv.cdf(y) * 2
    else:
        p = (1 - rv.cdf(y)) * 2
    print(f'pvalue {p:.3f}')
```

```python
pvar_test(sample, 9)
```

###       (    )


```python
def pmean_test(sample, mean0, alpha=0.05):
    s_mean = np.mean(sample)
    u_var = np.var(sample, ddof=1)
    n = len(sample)
    rv = stats.t(df=n-1)
    interval = rv.interval(1-alpha)

    t = (s_mean - mean0) / np.sqrt(u_var/n)
    if interval[0] <= t <= interval[1]:
        print('   ')
    else:
        print('   ')

    if t < 0:
        p = rv.cdf(t) * 2
    else:
        p = (1 - rv.cdf(t)) * 2
    print(f'pvalue {p:.3f}')
```

```python
pmean_test(sample, 130)
```

```python
t, p = stats.ttest_1samp(sample, 130)
t, p
```

### Wilcoxon signed  rank test


```python
training_rel = pd.read_excel('data/ch11_training_rel.xlsx')
toy_df = training_rel[:6].copy()
toy_df
```

### Chi- square test


```python
ad_df = pd.read_excel('data/ch11_ad.xlsx')
n = len(ad_df)
print(n)
ad_df
```

```python
ad_cross = pd.crosstab(ad_df['/'], ad_df[' '])
ad_cross
```

```python
ad_cross[''] / (ad_cross[''] + ad_cross[''])
```

```python
# ,    
n_not, n_yes = ad_cross.sum()
n_not, n_yes
```

```python
n_adA, n_adB = ad_cross.sum(axis=1)
n_adA, n_adB
```

```python
#expected frequency 
ad_ef = pd.DataFrame({'': [n_adA * n_yes / n,
                              n_adB * n_yes / n],
                      '': [n_adA * n_not / n,
                                   n_adB * n_not / n]},
                      index=['A', 'B'])
ad_ef
```

```python
y = ((ad_cross - ad_ef) ** 2 / ad_ef).sum().sum()
y
```

```python
rv = stats.chi2(1)
1 - rv.cdf(y)
```

```python
chi2, p, dof, ef = stats.chi2_contingency(ad_cross,
                                          correction=False)
chi2, p, dof
```
