## Магадлалын загвар

**магадалал(probability):**Тодохойгүй байдалд үйл явдлыг тайлбарлахын тулд ашигладаг. 

**магадлалын загвар(probability model):** Магадлалыг ашиглан санамсаргүй сонголт хийх

### Магадлалын үндэс

### Магадлалm тархалт

```python
dice = [1, 2, 3, 4, 5, 6]
prob = [1/21, 2/21, 3/21, 4/21, 5/21, 6/21]
prob
```

```python
np.random.choice(dice, p=prob)
```

```python
# 100 удаа шидэхэд 
num_trial = 100
sample = np.random.choice(dice, num_trial, p=prob)
sample
```

```python
# тархалтын давтамжын хүснэгт
freq, _ = np.histogram(sample, bins=6, range=(1, 7))
pd.DataFrame({'frequency':freq,
              'relative frequency':freq / num_trial},
              index = pd.Index(np.arange(1, 7), name='dice'))
```

```python
# 100 удаа шидэхэд гарах histogramaar
# шооны аль нүд нь хэд хэдэн удаа буух магадлалын тархац 
# гэтэл 100 удаа шоог шидэх харьцангүй давтамжаас гарсан магадлал ба бодитоор оноож өгсөн(prob)
# тархалт 2 ын хооронд зөрүү байгааг харж болно.
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
# Бодит магадлалын тархалтыг хэвтээ шугамаар харуулна
ax.hlines(prob, np.arange(1, 7), np.arange(2, 8), colors='gray')
# Бааран графикын [1.5, 2.5, ..., 6.5] т тэмдэглэ
ax.set_xticks(np.linspace(1.5, 6.5, 6))
# шооны утга нь [1, 2, 3, 4, 5, 6]
ax.set_xticklabels(np.arange(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')
plt.show()
```

```python
# шоог 1000 удаа шидэхэд
# энэ мэтчилэн шидэлтийн тоог ихэсгэх тусам харьцангүй давтамж нь бодит магадлал
# тархалттай ижил болж ирнэ.
# илүү дэлгэрэнгүйг animation folder chap 4 - oos
num_trial = 10000
sample = np.random.choice(dice, size=num_trial, p=prob)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample, bins=6, range=(1, 7), density=True, rwidth=0.8)
ax.hlines(prob, np.arange(1, 7), np.arange(2, 8), colors='gray')
ax.set_xticks(np.linspace(1.5, 6.5, 6))
ax.set_xticklabels(np.arange(1, 7))
ax.set_xlabel('dice')
ax.set_ylabel('relative frequency')
plt.show()
```

## статистик таамаглал ба магадлал

```python
# Түүвэр тус бүр санамсаргүй хувьсагч учир, түүврүүдийн дунджаар тооцоологдсон түүврийн дундаж нь 
# мөн санамсаргүй хувьсагч байна. 
sample_means = [np.random.choice(scores, 20).mean()
                for _ in range(10000)]

# 20 түүвэрийг сонгоно дундажийг нь 10000 удаа тооцоолж үзий
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
ax.hist(sample_means, bins=100, range=(0, 100), density=True)
# эх олонлогийн дундажийг босоо шугам хэлбэрээр харуулна
ax.vlines(np.mean(scores), 0, 1, 'gray')
ax.set_xlim(50, 90)
ax.set_ylim(0, 0.13)
ax.set_xlabel('score')
ax.set_ylabel('relative frequency')
plt.show()
# голын зураас нь эх олонлогийн дундаж 
# санамсаргүй сонголтын хувьд түүврийн дундажаар эх олонлогийн дундажийг тодорхойлж байна.
```
